```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('/content/drive/MyDrive/fear_greed_index.csv',
                 parse_dates=['date'],
                 dayfirst=True)

print(df.head())
print(df.dtypes)
```

```python
df.rename(columns={
    'timestamp': 'timestamp',
    'value': 'index_value',
    'classification': 'sentiment',
    'date': 'Date'
}, inplace=True)

df['timestamp'] = pd.to_datetime(df['timestamp'], unit='s')
```

```python
print(df.head())
print("\nData Types:\n", df.dtypes)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df['Date'], df['index_value'], label='Fear & Greed Index', color='blue')


ax.axhspan(0, 25, color='red', alpha=0.2, label='Extreme Fear')

ax.axhspan(75, 100, color='green', alpha=0.2, label='Extreme Greed')

ax.set_title('Fear & Greed Index with Sentiment Zones')
ax.set_xlabel('Date')
ax.set_ylabel('Index Value (0100)')
ax.legend()
plt.tight_layout()
plt.show()
```
